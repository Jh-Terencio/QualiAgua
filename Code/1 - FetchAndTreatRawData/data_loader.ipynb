# Data Loader para os dados do INEA
- Notebook responsável por baixar os dados brutos da qualidade da água de Lagoas, Rios e Baias disponibilizado pelo INEA

## Utils

### Importações

```python
import os
import requests
from pathlib import Path
from bs4 import BeautifulSoup
```

### Configurações

```python
page_url = "https://www.inea.rj.gov.br/ar-agua-e-solo/rios-baias-lagoas/"
DATA_PATH = Path("../../Data/RawData")
NEW_FILE_NAME = "WaterQualityRawData.xlsx"
```

## Code

```python
# Faz a requisição HTTP para obter o conteúdo da página
response = requests.get(page_url, timeout=30)
response.raise_for_status()
```

```python
# Analisa o conteúdo HTML da página
soup = BeautifulSoup(response.text, "html.parser")

link = soup.find(
    "a",
    string=lambda text: text and "Dados Brutos" in text
)


if not link:
    raise Exception("Link do arquivo não encontrado")
```

```python
# Constrói a URL completa do arquivo
file_url = link["href"]
file_name = file_url.split("/")[-1]

file_response = requests.get(file_url, stream=True)
file_response.raise_for_status()
```

```python
# Salva o arquivo localmente
file_path = os.path.join(DATA_PATH, NEW_FILE_NAME)

# Salva o arquivo localmente
with open(file_path, "wb") as f:
    for chunk in file_response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)

print(f"Download concluído: {file_path}")
```
